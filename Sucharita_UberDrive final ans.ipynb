## Q17. List the most frequent route(same pair of START & END points) taken by Uber drivers. 
## (3 points)

```python
j=uber.sort_values(by='START*', ascending=True).head()
k=uber.sort_values(by='STOP*', ascending=True).head()
pd.concat([j,k])
```

```python
loc1=uber[['START*']]
loc2=uber[['STOP*']]
pd.concat([loc1,loc2])
```

```python
loc1=uber[['START*']]
loc2=uber[['STOP*']]
pd.concat([loc1,loc2]).nunique()
```

```python
x=uber[['START*', 'MILES*']].groupby('START*').max().round()
y=uber[['STOP*', 'MILES*']].groupby('STOP*').max().round()
pd.concat([x,y])
```
